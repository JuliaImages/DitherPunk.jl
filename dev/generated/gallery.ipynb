The order of the Bayer-matrix can be specified through the parameter `level`,
which defaults to `1`.

```julia
img = gradient_image(100, 800)
dithers = [Gray.(bayer_dithering(img; level=level)) for level in 0:4]
mosaicview([img, dithers...])
```
